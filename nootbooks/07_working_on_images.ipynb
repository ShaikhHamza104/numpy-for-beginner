```python
import  numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Read image
img = cv2.imread('../images/img.png')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
# Flip image
flip = np.flip(img)
plt.imshow(flip)
```

```python
# clip image
plt.imshow(np.clip(img,0,100))
```

```python
# Convert BGR to RGB (Matplotlib expects RGB images)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Create negative image
negative_img = 255 - img_rgb

# Display the negative image
plt.imshow(negative_img)
plt.axis('off')  # Hide axes
plt.show()
```

```python
# trim
plt.imshow(img[100:500,50:900,:])
```

```python
# histogram
plt.hist(img.flatten(), bins=255)
```
